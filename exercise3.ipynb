Find the singular value decomposition of the following matrix:
$$A=
\begin{bmatrix}
    1    &0    &0   \\
    0    &-1   &0   \\
    0    &0    &-2
\end{bmatrix}
$$
Eigenvalues and eigenvectors of $AA^T$are:
$$AA^T = \begin{bmatrix}
    1    &0    &0   \\
    0    &-1   &0   \\
    0    &0    &-2
    \end{bmatrix}
    \begin{bmatrix}
    1    &0    &0   \\
    0    &-1   &0   \\
    0    &0    &-2
    \end{bmatrix}
    =
    \begin{bmatrix}
    1    &0    &0   \\
    0    &1   &0   \\
    0    &0    &4
    \end{bmatrix}
$$
 

$\lambda_1 = 4$, $\lambda_2=1$, $\lambda_3 = 1$, $v_1 = [0,0,1]^T$, $v_2 = [0,1,0]^T$, $v_3 = [1,0,0]^T$

Singular values are 

$\sigma_1 = \sqrt{4} = 2$, 

$\sigma_2 = \sqrt{1} = 1$, 

$\sigma_3 = \sqrt{1} = 1$


$\Sigma$ is a 3-by-3 matrix with the singular values on the diagonal.

$$
\Sigma =
    \begin{bmatrix}
    2    &0    &0   \\
    0    &1   &0   \\
    0    &0    &1
    \end{bmatrix} \\
$$

$$
V = [v_1, v_2, v_3] = \begin{bmatrix}
                        0    &0    &1   \\
                        0    &1    &0   \\
                        1    &0    &0
                        \end{bmatrix}
$$

$$u_1 = \frac{1}{\sigma_1}Av_1 = \begin{bmatrix} 0\\0\\-1\end{bmatrix}, \quad
  u_2 = \frac{1}{\sigma_2}Av_2 = \begin{bmatrix} 0\\-1\\0\end{bmatrix}, \quad
  u_3 = \frac{1}{\sigma_3}Av_3 = \begin{bmatrix} 1\\0\\0\end{bmatrix}$$
  
$$U = \begin{bmatrix}
                        0    &0     &1   \\
                        0    &-1    &0   \\
                        -1    &0    &0
                        \end{bmatrix}
$$

$$A = U\Sigma V^T = \begin{bmatrix}
                        0    &0     &1   \\
                        0    &-1    &0   \\
                        -1    &0    &0
                        \end{bmatrix}
                        \begin{bmatrix}
                        2    &0    &0   \\
                        0    &1   &0   \\
                        0    &0    &1
                        \end{bmatrix}
                        \begin{bmatrix}
                        0    &0    &1   \\
                        0    &1    &0   \\
                        1    &0    &0
                        \end{bmatrix}\\
$$


```python
np.array([[0,0,1],[0,-1,0],[-1,0,0]]).dot(np.array([[2,0,0],[0,1,0],[0,0,1]])).dot(np.array([[0,0,1],[0,1,0],[1,0,0]]))
```

```python
A4 = np.array([[1,0,0],
               [0,-1,0],
               [0,0,-2]
              ])

u, s, v = np.linalg.svd(A4)

print('U =')
u
print('S =')
s
print('V = ')
v

print("Test USV =")
smat = np.zeros(A4.shape)
smat[:len(s), :len(s)] = np.diag(s)

u.dot(smat).dot(v)

np.allclose(A4, np.dot(u, np.dot(smat, v)))
```

<a id='8'></a>

Suppose that the mortality for ages over 62 is decreasing at a rate of half a p ercent (0.005) a year over the next 40 years. What would the life expectancy coecient be in 2058? Use the table given in https://www.etk.fi/wp-content/uploads/EAK2018.pdf


$$EAL^{2018} = \sum_{x=62}^{n} 1,02^{-(x+0.5-62)} \frac{L_x^{2018}}{l_{62}^{2018}}$$

$$EAK^{2018} = \frac{EAL^{2009}}{EAL^{2018}}$$

$$EAK^{2058} = \frac{EAL^{2009}}{EAL^{2058}}$$


```python
df = pd.read_csv('/home/oscar/Dropbox/StochasticDemography/data/livslangdstabell09-18.csv')

df.head()
df.tail()
```

```python
#check 16,778288?, 17,458849
np.sum(df['val09'])
np.sum(df['val18'])
```

### Adding new $q^{2058}$ values

$$q^{2058}_x = q^{2018}_x \cdot 0.995^{40}$$

```python
r = 0.005
q = 1 - r

df['qx58'] = df['qx18'] * q**40
df.iloc[-1,9] = 1000
df.head()
df.tail()
```

### Computing new $l_x^{2058}$
$$l_{62}^{2058} = 1$$
$$l^{2058}_{x+1} = (1 - \frac{q^{2058}_{x}}{1000}) \cdot l_x^{2058}$$

```python
l = 1
l_values = [l]
for q in df['qx58']:
    l = (1 - q / 1000) * l
    l_values.append(l)

df['lx58'] = pd.Series(l_values)
df.head()
df.tail()
```

### Computing new $L_x$ values

$$L_x^{2058} = \frac{l_x^{2058} + l_{x+1}^{2058}}{2}$$

```python
ls = df['lx58'].values

Ls = (ls[:-1] + ls[1:])/2

df['Lx58'] = pd.Series(Ls)
df.iloc[-1, 11] = ls[-1] / 2
df.head()
df.tail()
```

$$EAL^{2058} = \sum_{x=62}^{n} 1,02^{-(x+0.5-62)} \frac{L_x^{2058}}{l_{62}^{2058}}$$

```python
l62 = df.loc[0,'lx58']
vals = []
for age, L in zip(df['age'], df['Lx58']):
    vals.append(1.02 ** -(age + 0.5 - 62) * L)

df['val_58'] = pd.Series(vals) / l62

df.head(20)
df.tail(19)
```

```python
EAL_58 = np.sum(df['val_58'])

EAL_09 = 16.778288

EAK_58 = EAL_09 / EAL_58

print("EAL 2058 = ", EAL_58)
print("EAK 2058 = ", EAK_58)
```

$$EAK^{2058} = \frac{EAL^{2009}}{EAL^{2058}} = \frac{16.778288}{18.433020596880397} = 0.9102299816688512$$

Calculate the dominant eigenvalue Î» for the Finnish fertility and mortality data 2015. Assuming the same fertility and mortality over the foreseeable future, what level of immigration would keep the population constant?

Start by reading the mortality data for Finland 1965-2015 and select the data for 2015.

```python
df_mor = pd.read_excel('/home/oscar/Dropbox/StochasticDemography/data/FinlMort65-15.xlsx',index=False)
df_mor = df_mor.iloc[:,:10]
df_mor15 = df_mor[df_mor['year'] == 2015]
df_mor15.head()
df_mor15.tail()
```

Then the data for fertality is loaded and the data for year 2015 is selected.

```python
df_fer = pd.read_excel('/home/oscar/Dropbox/StochasticDemography/data/FinlFert39-15.xlsx',index=False, skiprows=2)
df_fer15 = df_fer[df_fer['Year'] == 2015]
df_fer15.head(10)
df_fer15.tail(10)
```

The ASFR-column is the f-values in the Leslie-Matrix, that is the upper row.

```python
upper_row = df_fer15['ASFR'].values
upper_row
```

The lower diagonal contains the probabilities $s_x$ for survival into the next state.

$$s_x = \frac{l_{x+1}}{l_{x}}$$

```python
L = df_mor15['l(x)'].values
S = L[1:] / L[0:-1]
size = len(L)
size
```

There are as many states as there are ages. So the Leslie-Matrix will be a 111 by 111 matrix.

```python
A = np.zeros([size, size])

A[np.arange(size-1) + 1, np.arange(size-1)] = S
A[0, 12:len(upper_row) + 12] = upper_row
```

```python
r, c = 8,6
print("Upper left \n{}\n".format(A[:r,: c]))
print("Upper right \n{}\n".format(A[:r, -c:]))

print("Lower left \n{}\n".format(A[-r:, :c]))
print("Lower right \n{}\n".format(A[-r:, -c:]))

print("Upper row \n{}\n".format(A[0, 12:55]))
```

```python
eig = np.linalg.eigvals(A)

eig_real = [x.real for x in eig if x.imag==0 and x.real !=0]
```

```python
print("The real valued eigenvalues: ", eig_real)
```
